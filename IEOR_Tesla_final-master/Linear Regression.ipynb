```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data=pd.read_csv('df_lower_alldata.csv')
time=data.iloc[:,1]
data=data.drop(['Unnamed: 0','Interval Ending (AEST)'],axis=1)
time
```

```python
# Check the NaN values 
print("The count of NaN values is:", data.isnull().sum())
# Assume unnamed wind generataion farm initially not and fill the NaN value with zeros
data=data.fillna(0)
print("The count of NaN values is:", data.isnull().sum())
```

```python
#nomalize results 
from sklearn import preprocessing

scaler = preprocessing.StandardScaler()
scaled_df_test = scaler.fit_transform(data.iloc[:,0:19])

new=pd.DataFrame(scaled_df_test)
new
```

```python
# Split the dataset randomly into train and test 70% training and 30% validation 
X=new
Y=pd.DataFrame(data.iloc[:,-1])

x_train=X.iloc[0:15000,:]
x_test=X.iloc[15000:20778,:]
y_train=Y.iloc[0:15000]
y_test=Y.iloc[15000:20778]
```

```python
from sklearn import linear_model
clf = linear_model.LinearRegression() 
clf.fit(x_train, y_train)
y_train_pred = clf.predict(x_train)
y_test_pred = clf.predict(x_test)
print('Coefficients:', clf.coef_)
print('Intercepts:', clf.intercept_)
clf.score(x_train,y_train)
```

```python
#  Calculate 'Root Mean Square error' (RMSE) metric
def find_rmse (x,y):
    predicted = clf.predict(x) 
    rmse=np.mean((predicted-y)**2)**0.5 
    print(rmse)
print ("\n RMSE for training dataset is ") 
find_rmse(x_train,y_train)
print ("\n RMSE for testing dataset is ") 
find_rmse(x_test,y_test)
```
